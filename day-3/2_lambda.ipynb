```python
## Lambda functions are small anonymous functions defined using the lambda keyword. They can have any
## number of arguments but only one expression. They are commonly used for short operations or as arguments
## to higher-order functions.
```

```python
# syntax
# lambda arguments: expression
```

```python
addition = lambda a,b: a + b

print(addition(2,3))
```

```python
even_or_not = lambda num: num%2 ==0

even_or_not(12)
```

```python
addition1 = lambda a,b,c : a + b + c
addition1(2,3,4)
```
