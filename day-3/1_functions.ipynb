```python
def even_or_odd(num):
    """this function checks whether the function is odd or even"""
    if num%2 == 0:
        print("this number is even")
    else:
        print("this number is odd")
```

```python
even_or_odd(5)
```

```python
## function with multile params

def add(a,b):
    return a + b
result = add(1,2)
print(result)
```

```python
## default params

def greet(name='jatin'):
    print(f"hello {name}")

greet("kamal")
greet()
```

```python
## variable length arguments
## possitional and keyword

def print_numbers(*args):
    for numbers in args:
        print(numbers)

print_numbers(1,2,3,4,5,"jatin")
```

```python
## keyword argument

def print_details(**kwargs):
    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(name= "jatin",age = 20 , country = "India")
```

```python

def print_details(*args ,**kwargs):
    for val in args:
        print(f"Possitional arguments : {val}")

    for key, value in kwargs.items():
        print(f"{key}: {value}")

print_details(1,2,2,3,"kamal",3,name= "jatin",age = 20 , country = "India")
```

```python
## return statements

def multiply(a,b):
    return a*b,a

result,first_num = multiply(2,3)
print(result)
print(first_num)
```
